```python
options = Options()
driver = webdriver.Chrome()
driver.get(url)

buyer_names = []
comments = []

def find_reviews():
    s = ""
    buyers = driver.find_elements_by_class_name('middle')
    for name in buyers:
        s = name.text
        s = s.replace("Verified Purchase", "")
        s = s.replace("by", "")
        buyer_names.append(s)
        
    item_content = driver.find_elements_by_class_name('content')
    for i in range(1,len(item_content)):
        t = item_content[i].text
        comments.append(t)
        
def click_function():
    driver.find_element_by_xpath('/html/body/div[4]/div/div[8]/div[1]/div[2]/div/div/div[3]/div[2]/div/button[2]').click()
    time.sleep(5)

find_reviews()
click_function()
find_reviews()

df = pd.DataFrame({'Name': buyer_names,'Review Text': comments})
```

Scrape 80 items for a keyword search e.g., 'mobile covers'<br/>Note: Must use "click()" function of webElement, URL changing for pages is not allowed. The code must be generic and should work for every keyword, with proper checks in place. Bonus for also scraping country and Daraz Mall availability.

```python
query = 'mobile covers'
query = query.replace(' ', '+')
url = "https://www.daraz.pk/catalog/?q=" + query
```

```python
driver = webdriver.Chrome()
driver.get(url)

names = []
prices = []
reviews = []
country = []
ratings = []
availability = []

#                 FINDING PRODUCTS ON A SINGLE PAGE
def find_products():
    
    products = driver.find_elements_by_class_name('c5TXIP')

    for i in products:
        names.append(i.find_element_by_class_name('c16H9d').text)
        p = i.find_element_by_class_name('c13VH6').text
        p = p.replace('Rs.', '')
        prices.append(p)
        r = i.find_element_by_class_name('c3XbGJ').text
        r = r.replace('(', '').replace(')', '')
        reviews.append(r)
        country.append(i.find_element_by_class_name('c2i43- ').text)
        a = i.find_element_by_class_name('c3vCyH')
        try:
            a.find_element_by_tag_name('i').text
            availability.append("Yes")
        except NoSuchElementException:
            availability.append("No")

    for i in range(0, 40):
        product = driver.find_elements_by_class_name('c5TXIP')
        product[i].find_element_by_tag_name('a').click()
        time.sleep(3)
        if reviews[i] != ' ':
            r = driver.find_element_by_class_name('score').text
            r = r.replace('/5', '')
            ratings.append(r)
        driver.back()
        time.sleep(3)
        driver.refresh()
        
    if (len(names) < 80):
        click_function()
    
#                     CLICK FUNCTION ONTO NEXT PAGE
def click_function():
    time.sleep(3)
    driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[1]/div/div[1]/div[3]/div/ul/li[9]/a').click()
    time.sleep(5)
    find_products()
    driver.refresh()   

find_products()

if (len(names) == 80):
    df = pd.DataFrame({'Name':names,'Price':prices, 'Stars':ratings, 'Reviews':reviews, 'Country':country, 'Availability':availability})
```
